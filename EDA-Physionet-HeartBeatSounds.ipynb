Tranforming the data into structured folder.
Currently the folder contains the training folder which has three folders called a,b,c,d,e,f
Picking up all the files with (.wav) extension and moving into folders called "Abnormal" and "Normal" 

```python
import os
import shutil

# Root directory containing a, b, c, d... folders
root_dir = "training"

# Create global output folders
normal_out = os.path.join(root_dir, "normal")
abnormal_out = os.path.join(root_dir, "abnormal")

os.makedirs(normal_out, exist_ok=True)
os.makedirs(abnormal_out, exist_ok=True)

# Loop through each folder (a, b, c...)
for folder in os.listdir(root_dir):
    folder_path = os.path.join(root_dir, folder)

    # Skip non-folders
    if not os.path.isdir(folder_path):
        continue

    print(f"\nProcessing folder: {folder}")

    # Paths to text files
    normal_txt = os.path.join(folder_path, "records-normal.txt")
    abnormal_txt = os.path.join(folder_path, "records-abnormal.txt")

    # Load normal file list
    if os.path.exists(normal_txt):
        with open(normal_txt, "r") as f:
            normal_files = [line.strip() for line in f if line.strip()]
    else:
        normal_files = []

    # Load abnormal file list
    if os.path.exists(abnormal_txt):
        with open(abnormal_txt, "r") as f:
            abnormal_files = [line.strip() for line in f if line.strip()]
    else:
        abnormal_files = []

    # Move normal files
    for fname in normal_files:
        src = os.path.join(folder_path, fname)
        if os.path.exists(src):
            shutil.move(src, os.path.join(normal_out, fname))
            print(f"Moved NORMAL: {fname}")
        else:
            print(f"Missing normal file: {fname}")

    # Move abnormal files
    for fname in abnormal_files:
        src = os.path.join(folder_path, fname)
        if os.path.exists(src):
            shutil.move(src, os.path.join(abnormal_out, fname))
            print(f"Moved ABNORMAL: {fname}")
        else:
            print(f"Missing abnormal file: {fname}")

print("\nDone moving all files!")
```
