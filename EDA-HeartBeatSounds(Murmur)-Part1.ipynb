Visualization to Visualize the presence/occurence of nunber of times murmur present or absent

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv(file_path)

df.columns = df.columns.str.strip().str.lower()

labels = ['Absent', 'Present', 'Unknown']
counts = df['murmur'].str.strip().str.capitalize().value_counts().reindex(labels, fill_value=0)

# Step 4: Print counts for verification
print("Counts of Murmur Labels:")
print(counts)

plt.figure(figsize=(7,5))
ax = counts.plot(kind='bar', color=['green', 'red', 'gray'])

for i, val in enumerate(counts):
    ax.text(i, val + 5, f'{val/counts.sum()*100:.1f}%', ha='center', fontsize=12)

plt.title('Murmur Label Distribution (Including Unknowns)', fontsize=14)
plt.xlabel('Murmur Label', fontsize=12)
plt.ylabel('Number of Recordings', fontsize=12)
plt.xticks(rotation=0)
plt.grid(axis='y', linestyle='--', alpha=0.7)

plt.show()
```
